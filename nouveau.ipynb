```python
class Stimulus:
    def __init__(self, numero: str, temps_prst_apprent: int) -> None:
        self.numero = numero
        self.temps_prst_apprent = temps_prst_apprent
        self.temps_prst_reco = 1 - self.temps_prst_apprent
        self.statut : str = "non vu"
        self.correct_responses = self.correct_response()
    def le_sujet_repond(self) -> str:
        return(input(f"Already seen ? YN : {self.temps_prst_apprent}"))
    def correct_response(self)-> str:
        if self.statut == "non vu":
            self.correct_responses = "N"
        else :
            self.correct_responses = "Y"
        return self.correct_responses
             
        
class Experience:
    def __init__(self, liste_stimuli : list[Stimulus], tps_prst_general : int) -> None:
        self.liste_stimuli = liste_stimuli
        self.tps_prst_general = tps_prst_general

    def mise_a_jour_temps_prst_tous_les_stimuli(self)-> None:
        for stimulus in self.liste_stimuli:
            stimulus.temps_prst_apprent = self.tps_prst_general

    def deroulement_reco_un_stimulus(self, stimulus: Stimulus)-> None:
        reponse_du_sujet = stimulus.le_sujet_repond()
        if  reponse_du_sujet == stimulus.correct_response():
            self.tps_prst_general -= 50
            print(self.tps_prst_general)
            print(reponse_du_sujet, stimulus.correct_responses)
        else : 
            self.tps_prst_general += 50
            print(self.tps_prst_general)
            print(reponse_du_sujet, stimulus.correct_responses)
        self.mise_a_jour_temps_prst_tous_les_stimuli()

    def deroulement_expe_prst_chaque_stimuli(self)->None:
        for stimulus in self.liste_stimuli:
            while stimulus.statut != "vu deux fois":
                if stimulus.statut == "non vu":
                    print(stimulus.numero)
                    #print(stimulus.statut)########
                    self.deroulement_reco_un_stimulus(stimulus)
                    stimulus.statut = "vu"
                elif stimulus.statut == "vu":
                    print(stimulus.numero)
                    #print(stimulus.statut)#########
                    self.deroulement_reco_un_stimulus(stimulus)
                    stimulus.statut = "vu deux fois"
             
```

```python
visage_1 = Stimulus("Visage 1", 500)
visage_2 = Stimulus("Visage 2", 500)
visage_3 = Stimulus("Visage 3", 500)
pool_non_vus  = [visage_1, visage_2, visage_3]
expe_test = Experience(pool_non_vus, 500) 
expe_test.deroulement_expe_prst_chaque_stimuli()
```
